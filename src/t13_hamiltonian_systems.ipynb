## Task

1. Define a Hamiltonian function $H(q, p)$ (e.g., Mass on a Spring or Pendulum).
2. Derive the equations of motion: $\dot{q} = \partial H / \partial p$, $\dot{p} = -\partial H / \partial q$.
3. Simulate the system steps and plot trajectories in phase space.
4. Verify energy conservation by plotting $H(q(t), p(t))$ over time.
