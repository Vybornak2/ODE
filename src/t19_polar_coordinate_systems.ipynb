## Task

1. Define a system best formulated in polar coordinates (e.g., $\dot{r} = r(1-r^2), \dot{\theta} = 1$).
2. Transform coordinates $x = r \cos \theta, y = r \sin \theta$ symbolically if starting from Cartesian.
3. Solve the decoupled differential equations for $r(t)$ and $\theta(t)$.
4. Visualize the radial dynamics on a line and the global flow in the plane.
